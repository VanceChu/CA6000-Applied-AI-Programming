```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.arange(31)
print(rng)

rnd = np.random.randint(0,10,size=(3,rng.size))
print(rnd)

yrs = 1990 + rng
print(yrs)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.stackplot(yrs,rng+rnd,labels = ['Samsung','Apple','Xiaomi'])
ax.set_title('Market Share')
ax.legend(loc='upper left')
ax.set_ylabel('Total numbers (x 1000)')
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm

N_points = 100
x = np.linspace(-2, 2, N_points)
y = np.linspace(-2, 2, N_points)
X, Y = np.meshgrid(x, y)

Z = X**2 + Y**2        # function to plot
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(2, 2, 1, projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)

# Creating contour plot
plt.subplot(2, 2, 3)
plt.contourf(X, Y, Z, cmap=cm.coolwarm)

plt.show()
```

## Seaborn-Data Visual

```python
# [cite_start]Initialize the figure and axis objects using Matplotlib [cite: 984]
fig, ax = plt.subplots() 

# [cite_start]Set the title for the plot [cite: 984]
ax.set_title("Count of tables served by Day") 

# [cite_start]Define a custom order for the categorical x-axis [cite: 985]
days = ["Thur", "Fri", "Sat", "Sun"]

# [cite_start]Create a countplot: counts observations per category [cite: 986]
# x="day": Group data by day on the x-axis
# hue="time": Sub-group data by time using different colors
# order=days: Apply the custom order defined above
sns.countplot(x="day", hue="time", order=days, data=df) 

# [cite_start]Create a countplot with a specific color palette (green and yellow) [cite: 987]
sns.countplot(x="day", hue="time", order=days, data=df, palette=["green", "yellow"])
```

```python
# Create a figure with 1 row and 2 columns (subplots), setting size to 15x4
fig, axes = plt.subplots(1, 2, figsize=(15, 4))

# Define the order for x-axis categories
days = ["Thur", "Fri", "Sat", "Sun"]

# Plot 1: Countplot grouped by party 'size' on the first subplot (axes[0])
sns.countplot(x="day", hue="size", data=df, order=days, ax=axes[0])
axes[0].set_title("Tables served by Day and by party size")

# Plot 2: Countplot grouped by 'sex' on the second subplot (axes[1])
# Uses a dictionary to map specific colors to categories
sns.countplot(x="day", hue="sex", order=days, 
              palette=dict(Female="pink", Male="lightblue"), 
              data=df, ax=axes[1])
axes[1].set_title("Bill Payer")
```

```python
# 创建一个图形对象和坐标轴对象
fig, ax = plt.subplots()

# 设置图表的标题
ax.set_title("Histogram of Total Bill amounts") 

# 绘制直方图
# df['total_bill']: 指定要分析的数据列（总账单金额）
# color="lightgreen": 设置柱子的颜色为浅绿色
# bins=30: 将数据分为 30 个区间（箱子）来统计
sns.histplot(df['total_bill'], color="lightgreen", bins=30) 

# 添加平均值辅助线 (垂直线)
# df['total_bill'].mean(): 计算平均值位置
# linestyle="--": 线型为虚线
ax.axvline(df['total_bill'].mean(), color='red', linewidth=2, linestyle="--") 

# 添加中位数辅助线 (垂直线)
# df['total_bill'].quantile(q=0.5): 计算中位数位置 (50% 分位数)
# linestyle=":": 线型为点线
ax.axvline(df['total_bill'].quantile(q=0.5), color='blue', linewidth=2, linestyle=":") 
```
