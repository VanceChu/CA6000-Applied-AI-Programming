```python
import numpy as np
import time

# 创建一个包含 100 万个数字的数据
data_list = list(range(1000000))
data_arr = np.array(data_list)

# --- 方式 A: Python List (慢) ---
# 你必须写循环，逐个处理
t1 = time.time()
new_list = [x * 2 for x in data_list] 
print(f"Python List 耗时: {time.time() - t1:.5f} 秒")

# --- 方式 B: NumPy (快) ---
# 只要写 arr * 2，它会自动应用到所有元素
t2 = time.time()
new_arr = data_arr * 2 
print(f"NumPy Array 耗时: {time.time() - t2:.5f} 秒")
```

```python
import numpy as np

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# Example usage:
arr = np.array([3, 6, 8, 10, 1, 2, 15])
print("Original array:", arr.tolist())
sorted_arr = quick_sort(arr)
print("Sorted array:", sorted_arr.tolist())
```
